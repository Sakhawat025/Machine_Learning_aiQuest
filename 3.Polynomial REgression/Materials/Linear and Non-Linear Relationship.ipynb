# Now Sklearn Implementation

```python
#linear
from sklearn.linear_model import LinearRegression
df.head()
```

```python
reg1=LinearRegression()
reg1.fit(df[['x']], df[['y']])  #training
```

```python
reg1.score(df[['x']], df[['y']])  #training score
```

```python
x=df[['x']] #2d
y=df['y']
y.shape
```

```python
from sklearn.metrics import r2_score   # imports R (coefficient of determination) metric

r2_score(y, reg1.predict(x))  
# calculates R score  measures how well the model explains variance in y
# Formula: R = 1  ((yi  i) / (yi  ))
```

```python
#linear
plt.figure(figsize=(16,10))
plt.scatter(df['x'],
           df['y'],
           label='Data with Linear Relationship',
           c=df['y'],
           cmap='magma',
           s=60,
           alpha=.90,
           edgecolor='k')

plt.scatter(df[['x']], reg1.predict(df[['x']]), color='blue', marker='*', s=50, label='Best fit Line')
plt.xlabel('x-Axis Data', fontsize=10)
plt.ylabel('y-Axis Data', fontsize=10)
plt.title('linear Relationship', fontsize=15, weight='bold')
plt.legend()
```

# Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=6)      # polynomial regression with degree n
X_poly = poly.fit_transform(df2[['x']])
```

```python
X_poly.shape
```

```python
X_poly
```

```python
pd.DataFrame(X_poly).head()
```

```python
df2.x.head()
```

# Polynomial Regression = Poly Features + Linear Regresion

```python
reg_poly = LinearRegression()
```

```python
reg_poly.fit(X_poly, df2.y) # all data
reg_poly.score(X_poly, df2.y)
```

```python
plt.figure(figsize=(16,10))
plt.scatter(df2['x'],
           df2['y'],
           label='Data with Non-linear Relationship',
           c=df2['y'],
           cmap='magma',
           s=62,
           alpha=.89,
           edgecolor='k')

plt.plot(df2.x, 2*np.sin(df2.x), color='red', label='True Non_linear Relationship')
plt.xlabel('x-Axis Data', fontsize=12)
plt.ylabel('y-Axis Data', fontsize=12)
plt.title('Non_Linear Relationship', fontsize=15, weight='bold')
plt.legend()
```
